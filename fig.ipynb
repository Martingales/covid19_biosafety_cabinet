```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(index = ['Coronavirus', 'Influenza A','Staphylococcus aureus','Mycoplasma spp.','Aspergillus flavus'],columns=['Integral Dose '])
```

```python
df['Integral Dose'] = np.array([3200,1900,6600,8400,34900])
```

```python
[np.arange(50,251,1) *i for i in df['Integral Dose']];
```

```python
fig,ax= plt.subplots(figsize=(10,6))
cols = ['Navy', 'Green','lightblue','Orange','purple']
for i,data in enumerate([np.arange(50,251,1) *i for i in df['Integral Dose']][::-1]):
    ax.plot(np.arange(50,251,1),data/(1000*60),label=df.index[len(df.index)-1-i],lw=7,color=cols[i])
    
ax.legend(bbox_to_anchor=[1, 0.7],edgecolor='w',fontsize=12)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)
#plt.yscale('log')
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(color='lightgrey', linestyle='-', linewidth=2)
ax.set_ylabel('Time to 90% sterilization (min.)',fontsize=16)
ax.set_xlabel(r'UV intensity $(\mu W\,cm^{-2})$',fontsize=16)
plt.tight_layout()
#plt.savefig('BSC_organisms.pdf')
```
